```python
# -*- coding: utf-8 -*-
"""Jhinjer - BBIM502

Automatically generated by Colab.

Original file is located at
    https://colab.research.google.com/drive/1M5m-Ju9VJuEIdD-cs5OVZbyjbZOFY6s4
"""

import os
import datetime

def add_room(rooms):
    try:
        room_number = int(input("Enter room number to add: "))
        if room_number in rooms:
            print("Room already exists.")
        else:
            room_type = input("Enter room type (Single/Double): ")
            price = float(input("Enter room price per night: "))
            rooms[room_number] = {'type': room_type, 'price': price, 'allocated': False, 'customer': None}
            print("Room added successfully.")
    except ValueError as e:
        print(f"Error: {e}")

def delete_room(rooms):
    try:
        room_number = int(input("Enter room number to delete: "))
        if room_number in rooms:
            del rooms[room_number]
            print("Room deleted successfully.")
        else:
            print("Room not found.")
    except ValueError as e:
        print(f"Error: {e}")

def display_rooms(rooms):
    if rooms:
        for room, details in rooms.items():
            print(f"Room {room}: Type: {details['type']}, Price: {details['price']}, Allocated: {details['allocated']}")
    else:
        print("No rooms available.")

def allocate_room(rooms):
    try:
        room_number = int(input("Enter room number to allocate: "))
        if room_number in rooms and not rooms[room_number]['allocated']:
            customer_name = input("Enter customer name: ")
            rooms[room_number]['allocated'] = True
            rooms[room_number]['customer'] = customer_name
            print("Room allocated successfully.")
        else:
            print("Room is either not available or already allocated.")
    except ValueError as e:
        print(f"Error: {e}")

def display_allocations(rooms):
    allocated_rooms = {k: v for k, v in rooms.items() if v['allocated']}
    if allocated_rooms:
        for room, details in allocated_rooms.items():
            print(f"Room {room} is allocated to {details['customer']}.")
    else:
        print("No rooms allocated.")

def bill_and_deallocate(rooms):
    try:
        room_number = int(input("Enter room number to bill and deallocate: "))
        if room_number in rooms and rooms[room_number]['allocated']:
            nights = int(input("Enter number of nights stayed: "))
            bill = rooms[room_number]['price'] * nights
            print(f"Customer {rooms[room_number]['customer']} is billed ${bill:.2f}.")
            rooms[room_number]['allocated'] = False
            rooms[room_number]['customer'] = None
            print("Room deallocated successfully.")
        else:
            print("Room is either not available or not allocated.")
    except ValueError as e:
        print(f"Error: {e}")

def save_allocation_to_file(rooms):
    try:
        with open("LHMS_Studentid.txt", "w") as file:
            for room, details in rooms.items():
                file.write(f"Room {room}: {details}\\n")
        print("Room allocation details saved to file.")
    except IOError as e:
        print(f"File error: {e}")

def show_allocation_from_file():
    try:
        if os.path.exists("LHMS_Studentid.txt"):
            with open("LHMS_Studentid.txt", "r") as file:
                print(file.read())
        else:
            print("File not found.")
    except IOError as e:
        print(f"File error: {e}")

def backup_allocation_file():
    try:
        if os.path.exists("LHMS_Studentid.txt"):
            timestamp = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
            backup_file = f"LHMS_Studentid_Backup_{timestamp}.txt"
            with open("LHMS_Studentid.txt", "r") as original_file:
                content = original_file.read()
            with open(backup_file, "w") as backup:
                backup.write(content)
            open("LHMS_Studentid.txt", "w").close()  # Clear the original file
            print(f"Backup created successfully as {backup_file}.")
        else:
            print("Original file not found.")
    except IOError as e:
        print(f"File error: {e}")

def main():
    rooms = {}
    while True:
        print("\nHotel Management Menu")
        print("1. Add Room")
        print("2. Delete Room")
        print("3. Display Room Details")
        print("4. Allocate Room")
        print("5. Display Room Allocation Details")
        print("6. Billing & De-Allocation")
        print("7. Save Room Allocation to File")
        print("8. Show Room Allocation from File")
        print("9. Backup Room Allocation File")
        print("0. Exit Application")

        try:
            choice = int(input("Enter your choice: "))

            if choice == 1:
                add_room(rooms)
            elif choice == 2:
                delete_room(rooms)
            elif choice == 3:
                display_rooms(rooms)
            elif choice == 4:
                allocate_room(rooms)
            elif choice == 5:
                display_allocations(rooms)
            elif choice == 6:
                bill_and_deallocate(rooms)
            elif choice == 7:
                save_allocation_to_file(rooms)
            elif choice == 8:
                show_allocation_from_file()
            elif choice == 9:
                backup_allocation_file()
            elif choice == 0:
                print("Exiting application. Goodbye!")
                break
            else:
                print("Invalid choice. Please try again.")

        except ValueError:
            print("Please enter a valid numeric choice.")

if __name__ == "__main__":
    main()
```
